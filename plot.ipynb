```python
import matplotlib.pyplot as plt
import numpy as np
import pandas
import seaborn as sns
import scipy.stats as st

path = 'logs/csil/AdroitHandHammer-v1/d3/s1/progress.csv'
file = pandas.read_csv(path)
```

```python
returns = file['Eval/ReturnMean']
ma_returns = file['Eval/ReturnMean'].rolling(window=5).mean()
steps = (file['Epoch'] + 1) * 10000

plt.plot(steps, returns)
plt.plot(steps, ma_returns)
plt.axhline(16800, color='r', ls=':')
plt.title("Returns")
# plt.savefig('temp')
plt.show()
```
