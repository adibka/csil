```python
import pandas as pd

def bc_runs(env, demo, max_steps=2):
    
    tresholds = {
        'ant': 4700,
        'walker': 5200,
        'humanoid': 5300,
        'cheetah': 5098,
        'hopper': 3500,
        'hammer': 16800,
    }
    expert_treshold = tresholds[env]
    steps = pd.read_csv(f"bc/{env}/d{demo}/s0/progress.csv")['Steps'][:max_steps] // 1000
    steps[0] = 0.0
    r0 = pd.read_csv(f"bc/{env}/d{demo}/s0/progress.csv")['Eval/ReturnMean'][:max_steps]
    r1 = pd.read_csv(f"bc/{env}/d{demo}/s1/progress.csv")['Eval/ReturnMean'][:max_steps]
    r2 = pd.read_csv(f"bc/{env}/d{demo}/s2/progress.csv")['Eval/ReturnMean'][:max_steps]
    r3 = pd.read_csv(f"bc/{env}/d{demo}/s3/progress.csv")['Eval/ReturnMean'][:max_steps]
    r4 = pd.read_csv(f"bc/{env}/d{demo}/s4/progress.csv")['Eval/ReturnMean'][:max_steps]
    
    data = {}
    data = {'r0': r0,
           'r1': r1,
           'r2': r2,
           'r3': r3,
           'r4': r4,
        }
    df = pd.DataFrame(data=data) / expert_treshold
    df['mean'] = df.mean(axis=1)
    df['std'] = df.std(axis=1)
    df['steps'] = steps
    df = df[1:].reset_index(drop=True)
    df.to_csv(f'bc-{env}-{demo}.csv', index=False)
    return df
```

```python
for env in ['ant', 'hopper', 'humanoid', 'walker', 'cheetah', 'hammer']:
    for demo in [3, 10]:
        bc_runs(env, demo)
```
